```python
# 16.Vocabulary Builder: Read a text file and create a dictionary where keys are words and values are how many times each word appears (Word Frequency).
with open('file3.txt.','r') as file :
    text = file.read()
    words = text.lower().split()
    word_count = {}
    for word in words :
        word = word.strip(".,!?\"'")
        if word in word_count:
            word_count[word] +=1
        else:
            word_count[word] = 1
    print (word_count)
```
