```python
# 16.Abstract Base Classes (ABC): Research the abc module. Create a class that forces all children to implement a specific method (e.g., every Shape must have an area() method).
from abc import ABC,abstractmethod
class shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

```python
class square(shape):
    def area(self,l,b):
        self.l = l 
        self.b = b
        area = self.l * self.b
        return area
    def features(self):
        print('Square has four sides')
```

```python
s1 =square()
```

```python
s1.area(3,3)
```

```python
s1.features()
```
