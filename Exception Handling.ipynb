```python
#Nested Try-Except: Put a try-except block inside another except block to try a backup solution if the first error-handling fails.
a = int(input())

try:
    result = 10/a
    print(result)
except ZeroDivisionError:
    print(ZeroDivisionError)
    try:
        result = 10/a   # backup solution
        print("Backup result:", result)
    except Exception as e:
        print("Backup also failed:", e)
except ValueError:
    print("Invaid Input! Please enter a number")
```
