```python
# Area Calculator: Create a class Circle that takes radius. Add a method calculate_area() (use 3.14 * r * r).
class Circle():
    def __init__(self,radius):
        self.radius = radius
    def calculate_area(self):
        area = 3.14 * self.radius * self.radius
        return area
```

```python
# List of Objects: Create a loop that generates 5 car objects with different names and stores them in a list.
class Car:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Car Name: {self.name}"



car_list = []


for i in range(5):
    car_name = input(f"Enter name of car {i+1}: ")
    car = Car(car_name)
    car_list.append(car)


for car in car_list:
    print(car)
```

```python
# Method Interaction: Create a User class and a Post class. Let the User object have a method create_post() that generates a Post object.
class Post:
    def __init__(self, content, author):
        self.content = content
        self.author = author

    def __str__(self):
        return f"{self.author} posted: {self.content}"


class User:
    def __init__(self, username):
        self.username = username

    def create_post(self, content):
        return Post(content, self.username)
```

```python
user1 = User("Bhavesh")
```

```python
post1 = user1.create_post("Hello, this is my first post!")
```

```python
post1.author
```

```python
post1.content
```
